**SELLING OF PRODUCTS FROM THE BAKERY INVENTORY AND PRINTING BILLS TO THE CUSTOMERS**

```python
import time
lst=[]
 
print("---------------------------------------------------------")
print("ETG - BAKERY INVENTORY MANAGEMENT SYSTEM")
print("BY KARKAVELRAJA J")
print("---------------------------------------------------------")
print(time.ctime())
print("---------------------------------------------------------")
 
user_input=int(input("ENTER THE NUMBER OF PRODUCTS : "))
print("---------------------------------------------------------")
for i in range(user_input) :
  print("DETAILS OF PRODUCT ",i+1)
  print("---------------------------------------------------------")
  product_id=input("ENTER THE PRODUCT ID : ")
  product_quan=int(input("ENTER THE QUANTITY OF PRODUCT : "))
 
  
  if product_id in record_dictionary.keys() :
    print("ID OF THE PRODUCT       : ",product_id)
    print("NAME OF THE PRODUCT     : ",record_dictionary[product_id]["product_name"])
    print("CATEGORY OF THE PRODUCT : ",record_dictionary[product_id]["product_category"])
    print("PRICE OF THE PRODUCT    : ",record_dictionary[product_id]["product_price"])
    print("QUANTITY OF THE PRODUCT : ",product_quan)
    record_dictionary[product_id]["product_quantity"]=record_dictionary[product_id]["product_quantity"] - (product_quan)
    amount=record_dictionary[product_id]["product_price"]*product_quan
    print("---------------------------------------------------------")
    print("AMOUNT OF THE PRODUCT   : ",amount)
    print("---------------------------------------------------------")
    lst.append(amount)
 
  else:
    print("---------------------------------------------------------")
    print("INVALID PRODUCT ID")
    print("---------------------------------------------------------")
 
print("---------------------------------------------------------")
print("TOTAL PRICE OF THE PRODUCT: ",sum(lst))
print("---------------------------------------------------------")
 
print("************THANK**YOU***********VISIT**AGAIN************")
print("---------------------------------------------------------")
```
